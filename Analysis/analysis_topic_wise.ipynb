```python
import pandas as pd 
```

```python
df = pd.read_csv("graph_SPORTS.csv")
```

```python
print(df.head())
```

```python
df.sort_values('Weight', ascending = False, inplace = True)
```

```python
df = df[:50]
```

```python
import networkx as nx 
G = nx.Graph()
```

```python
for i in range(len(df)):
    G.add_edge(df.iloc[i]['Source'],df.iloc[i]['Source'],weight = df.iloc[i]['Weight'])
```

```python
degree_centrality = nx.degree_centrality(G)
```

```python
Keymax = max(zip(degree_centrality.values(), degree_centrality.keys()))[1]
print(Keymax)
print(degree_centrality[Keymax])
```

```python
eigen_centrality = nx.eigenvector_centrality(G)
print(Keymax)
print(degree_centrality[Keymax])
```

```python
import networkx.algorithms.community as nx_comm
from networkx.algorithms import community
import community as community_louvain
import matplotlib.pyplot as plt
import matplotlib.cm as cm


partition = community_louvain.best_partition(G, weight = "None")
# draw the graph
pos = nx.spring_layout(G)
# color the nodes according to their partition
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40,
                       cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```
